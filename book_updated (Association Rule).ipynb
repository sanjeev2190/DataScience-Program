```python
import pandas as pd 
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
book=pd.read_csv("book.csv")
book
```

```python
frequent_itemset=apriori(book,min_support=0.1,use_colnames=True)
frequent_itemset
```

```python
rules=association_rules(frequent_itemset,metric="lift",min_threshold=1)
rules
```

```python
#An leverage value of 0 indicates independence. Range will be [-1 1]
#A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]
```

```python
rules.sort_values("lift",ascending=False).head(20)
```

```python
rules[rules.lift>1]
```

```python
#change the min_support value 0.1 to 0.2
#confidence value will not change bcz it will not availabe in a apriori and mlextend library
frequent_itemset=apriori(book,min_support=0.2,use_colnames=True)
frequent_itemset
```

```python
#change The thresold value 1 to 1.3

rules=association_rules(frequent_itemset,metric="lift",min_threshold=1.3)
rules
```

```python
rules.sort_values("lift",ascending=False).head(20)
```

```python
rules[rules.lift>1]
```
