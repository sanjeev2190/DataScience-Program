```python
import pandas as pd 
from mlxtend.frequent_patterns import association_rules,apriori
from mlxtend.preprocessing import TransactionEncoder
```

```python
mymovies=pd.read_csv("my_movies.csv")
mymovies
```

```python
mymovies.isnull().sum()
```

```python
mymovies.info()
```

```python
mymovies.V4.mode()
```

```python
mymovies=mymovies.drop(columns=["V1","V2","V3","V4","V5"],axis=1)
```

```python
mymovies
```

```python
#apriori Algorithm
#frequent_itemsets=apriori(mymovies,min_support=0.1,max_lan=3,use_colnames=True)
frequent_itemsets=apriori(mymovies,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
#confidence value will not change bcz it will not availabe in a apriori and mlextend library
```

```python
rules.sort_values('lift',ascending = False).head(20)
```

```python
rules[rules.lift>1]
```

```python
# i will change min_support value is 0.1 to 0.2 then output
frequent_itemsets=apriori(mymovies,min_support=0.2,use_colnames=True)
frequent_itemsets
```

```python
# in ass0ciation rule confidence function is used instead of lift
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=1)
rules
```

```python
rules.sort_values('confidence',ascending = False).head()
```

```python
rules[rules.lift>1]
```
